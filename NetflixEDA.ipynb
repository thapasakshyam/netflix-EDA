# **Exploratory Data Analysis of Netflix Data**

The dataset that is used in this project was sourced from [Kaggle.](https://www.kaggle.com/datasets/rahulvyasm/netflix-movies-and-tv-shows)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv", encoding = "latin-1")
df.head(10)
```

The data contains unnecessary and empty columns, so it has been dropped.

```python
df.columns
col_to_drop = df.columns[12:26]
df.drop(col_to_drop, axis = 1, inplace = True)
df.tail()
```

```python
df.info()
```

```python
type_counts = df['type'].value_counts()

plt.pie(type_counts, labels=type_counts.index, autopct="%1.1f%%")
plt.title("Distribution of Content Type")
plt.show()
```

There are few entries in the 'Country' column which have 'Nan' value so I am going to chane it to 'Unknown' rather than dropping it. Country represents where the TV shows and Movies have been produced.

```python
df['country'].isna().sum()
```

```python
df['country'].fillna('Unknown')
```

```python
country_counts = df['country'].value_counts().nlargest(10)
country_counts
```

```python
for i, v in enumerate(country_counts.values):
  plt.text(i, v + 0.1, str(v), ha='center', fontsize=8)

plt.bar(country_counts.index, country_counts.values)
plt.xlabel("Country")
plt.ylabel("Number of Conten")
plt.title("Top 10 Countries for Content Availability on Netflix")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

Dropping the empty values from 'date_added' column and splitting the dates into month and year.

```python
df['date_added'].isna().sum()
```

```python
df['date_added'].dropna()
```

```python
split_dates = df["date_added"].str.split(" ")

df["added_month"] = split_dates.str[0]
df["added_year"] = split_dates.str[2]
df.head()
```

```python
monthly_content_counts = (df.groupby(["added_month", "type"]).size().to_frame(name='count').unstack().dropna())

monthly_content_counts.plot(kind="line")
plt.xlabel("Month")
plt.ylabel("Content Count")
plt.title("Netflix Content Added by Month")
plt.xticks(rotation=45, ha = "right")
plt.legend(title="Content Type")
plt.tight_layout()
plt.figure(figsize=(12,12))
plt.show()

```
